```python
import pandas as pd
```

```python
df = pd.read_excel("data.xlsx")
df
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged_df = pd.concat([df,dummies],axis="columns")
merged_df
```

```python
final = merged_df.drop(['town','west windsor'], axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
x = final.drop('price',axis = 'columns')
x
```

```python
y = final.price
y
```

```python
model.fit(x,y)
```

home with 2800 area in robbinsville

```python
model.predict([[2800,0,1]])
```

home with 3400 area in west windsor

```python
model.predict([[3400,0,0]])
```

model prediction accuracy

```python
model.score(x,y)
```

# Exercise Car Price 

predict price mercedez benz that is 4 yr old with mileage 45000
predict price BMW that is 7 yr old with mileage 86000
the score (accuracy) of your model

```python
df = pd.read_excel("car-price.xlsx")
df
```

```python
dummies = pd.get_dummies(df['car model'])
dummies
```

```python
merged_df = pd.concat([df, dummies],axis='columns')
merged_df
```

```python
final = merged_df.drop(['car model','Mercedez'],axis='columns')
final
```

```python
x = final.drop('sell price', axis='columns')
x
```

```python
y = final['sell price']
y
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
model.predict([[45000,4,0,0]])
```

```python
model.predict([[86000,7,0,1]])
```

```python
model.score(x,y)
```
