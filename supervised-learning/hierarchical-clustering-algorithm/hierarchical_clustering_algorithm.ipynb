```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')
```

**Creating an Artificial Dataset**

```python
example_data = make_blobs(n_samples=300, n_features=2,
                         centers=3, cluster_std=1.3, random_state=42)
```

```python
dots = example_data[0]
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendogram_example = sch.dendrogram(sch.linkage(dots, method='ward'))
```
