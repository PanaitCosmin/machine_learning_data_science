```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_excel("price_home.xlsx")
df
```

```python
import math
median_bedrooms = math.floor(df.bedrooms.median())
median_bedrooms
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms','age']], df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

predict price for home with 3000 area 3 bedrooms and 40 y/d

```python
reg.predict([[3000,3,40]])
```

```python
reg.predict([[3000,4,15]])
```

Exercise(salary): 2yr exp, 9 test score, 6 interview score
                  12yr exp, 10 test score, 10 interview score

```python
df = pd.read_excel("hiring.xlsx")
df
```

```python
df.experience = df.experience.fillna("zero")
```

```python
df
```

```python
from word2number import w2n
```

```python
df.experience = df.experience.apply(w2n.word_to_num)
```

```python
df
```

```python
test_median = math.floor(df.test_score.median())
df.test_score = df.test_score.fillna(test_median)
```

```python
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['experience','test_score','interview_score']],df.salary)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2, 9, 6]])
```

```python
reg.predict([[12, 10, 10]])
```
